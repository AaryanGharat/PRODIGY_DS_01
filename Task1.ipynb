**TOTAL POPULATION OVER YEAR**


```python
years = df.columns[4:]  
population =df.iloc[:, 4:].sum()

plt.figure(figsize=(12, 13))
plt.bar(years, population, color='blue')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Total Population Over the Years ')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Top 30 Countries with Highest Population in 2022**

```python
countries = df['Country Code'].unique()
latest_population = [(country, df[df['Country Code'] == country]['2022'].values[0]) for country in countries]

sorted_countries = sorted(latest_population, key=lambda x: x[1], reverse=True)

plt.figure(figsize=(12, 10))

for i, (country, population) in enumerate(sorted_countries[:30]):
    plt.bar(country, population, color='skyblue')

plt.xlabel('Population (in billions)')
plt.title('Top 30 Countries with Highest Population (2022)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```

**Top 30 Counties with lowest Growth  in 2022**

```python
plt.figure(figsize=(12, 10))

for i, (country, population) in enumerate(sorted_countries[234:]):
    plt.bar(country, population, color='skyblue')

plt.ylabel('Population (in billions)')
plt.title('Top 30 Countries with Lowest Population (2022)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```

**Top 30 Countries with Highest Growth Percentage from 1960 to 2022**

```python
selected_years = ['1960','2022']

percentage_growths = []
for country in countries:
    country_data = df[df['Country Code'] == country][selected_years].values.flatten()
    growth_percentage = ((country_data[-1] - country_data[0]) / country_data[0]) * 100
    percentage_growths.append((country, growth_percentage))

sorted_countries = sorted(percentage_growths, key=lambda x: x[1], reverse=True)

plt.figure(figsize=(12, 10))

for i, (country, growth_percentage) in enumerate(sorted_countries[:30]):
    plt.bar(country, growth_percentage, color='skyblue')

plt.xlabel('Percentage Growth')
plt.title('Top 30 Countries with Highest Percentage Growth of Population (1960-2022)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```

**Top 30 Countries with Lowest Growth Percentage from 1960 to 2022**

```python
percentage_growths = []
for country in countries:
    country_data = df[df['Country Code'] == country][selected_years].values.flatten()
    growth_percentage = ((country_data[-1] - country_data[0]) / country_data[0]) * 100
    percentage_growths.append((country, growth_percentage))

sorted_countries = sorted(percentage_growths, key=lambda x: x[1], reverse=True)

plt.figure(figsize=(12, 10))

for i, (country, growth_percentage) in enumerate(sorted_countries[234:]):
    plt.bar(country, growth_percentage, color='skyblue')

plt.xlabel('Percentage Growth')
plt.title('Top 30 Countries with Lowest Percentage Growth of Population (1960-2022)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```
